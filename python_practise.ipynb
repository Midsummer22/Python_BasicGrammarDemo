```python
"""
类内置方法(又称为魔术方法);
__init__(),__str__(),__lt__(),__le__(),__eq__()
"""

# __str__()方法,__lt__()方法；
class Student:
    m_Name = 0
    m_Age = 0
    def __init__(self,name,age):
        self.m_Name = name
        self.m_Age = age

    def __str__(self):              # 有点像Cpp里的运算符重载，使得第22行可以直接打印当前对象;
        return f"Student类对象，name = {self.m_Name},age = {self.m_Age}"

    def __lt__(self, other):       # 类似于Cpp中"<"运算符的重载;
        return self.m_Age < other.m_Age

    def __le__(self,other):       # 注意运算符重载函数指定形式为魔术方法，类似于Cpp中的“operator”关键字;
        return self.m_Age <= other.m_Age

    def __eq__(self, other):       # operator==
        return self.m_Name == other.m_Name

student1 = Student("嘉文IV",24)
student2 = Student("赵信",25)
print(student2)                     # 如果__str__()方法未创建，则打印当前对象的内存地址;
print(student1 < student2)
print(student2 <= student1)
print(student1 >= student2)
print(f"嘉文年龄比赵信小？:{student1 < student2}")
print(f"当前学生是否重名:{student1 == student2}")
```

```python
"""
single-inheritance;
"""

# 单继承;
class Phone:
    IMEI = 0
    producer = "HuaWei"

    def call_by_4g(self):
        print("4g通话;")
        return

class Phone2022(Phone):         # 类后面括号用以表继承;
    face_id = "10001"

    def call_by_5g(self):
        print("2022年实现新功能:5g通话;")
        return

new_phone = Phone2022()
new_phone.IMEI = "HuaWei_new_phone"
print(f"新手机的厂商是:{new_phone.producer}")
print(f"新手机的面部ID为:{new_phone.face_id}")
new_phone.call_by_5g()
```

```python
"""
multi-inheritance;
"""

class Phone:
    IMEI = 0
    producer = 0
    def call_by_5g(self):
        print("5g通话;")
        return

class NFCReader:
    nfc_type = "the 5th grade"
    producer = "HuaWei"
    def read_card(self):
        print("读取NFC卡;")
    def write_card(self):
        print("写入NFC卡;")

class RemoteControl:
    rc_type = "红外遥控;"
    def control(self):
        print("红外遥控开启;")

class MyPhone(Phone,NFCReader,RemoteControl):
    pass        # pass:补全语法，实际此处为空;

myphone = MyPhone()
print(f"手机编号为{myphone.IMEI}")
print(f"手机NFC厂商:{myphone.producer}")        # 按照MyPhone中的参数继承顺序，优先第一个继承的变量;
myphone.control()
```

```python
# 函数形参及返回值的类型注解;
def add(x:int,y:int) -> int:
    return x + y

result = add(1,5)
print(result)
```

```python
# union联合注解;
from typing import Union

new_list:list[Union[str,int]] = [22,25,"Cpp","SLAM","3D Reconstruction","JavaScript","Python","Deep Learning"]
my_dict:dict[str,Union[str,int]] = {"Cpp":"Best Language in Computer,in my opinion;","Pytorch":1}   # Union联合注解：字典里的value项不是str就是int;
print(new_list,my_dict)
```

```python
# union联合注解演示;
from typing import Union
# 给数据类型注解;
new_list:list[Union[str,int]] = [22,25,"Cpp","SLAM","3D Reconstruction","JavaScript","Python","Deep Learning"]
my_dict:dict[str,Union[str,int]] = {"Cpp":"Best Language in Computer,in my opinion;","Pytorch":1}   # Union联合注解：字典里的value项不是str就是int;

# 给函数注解;
def func(data1:Union[int,float],data2:Union[int,float]) -> Union[int,float]:
    return data1 + data2
print(func(2,5))
```
